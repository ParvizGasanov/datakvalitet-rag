```python
df_clean = df.copy()
```

```python
text_columns = ["director", "cast", "country", "date_added", "rating", "duration"]

df_clean[text_columns] = df_clean[text_columns].fillna("Unknown")
```

```python
df_clean.isnull().sum()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
for col in df_clean.select_dtypes(include="object").columns:
    df_clean[col] = df_clean[col].str.strip()
```

```python
df_clean["rag_text"] = (
    "Title: " + df_clean["title"] + "\n" +
    "Type: " + df_clean["type"] + "\n" +
    "Director: " + df_clean["director"] + "\n" +
    "Cast: " + df_clean["cast"] + "\n" +
    "Country: " + df_clean["country"] + "\n" +
    "Release year: " + df_clean["release_year"].astype(str) + "\n" +
    "Rating: " + df_clean["rating"] + "\n" +
    "Duration: " + df_clean["duration"] + "\n" +
    "Categories: " + df_clean["listed_in"] + "\n" +
    "Description: " + df_clean["description"]
)

df_clean[["title", "rag_text"]].head()
```

```python
df_clean.to_csv("../data/cleaned_data.csv", index=False)
```
